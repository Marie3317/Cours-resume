```python
# Types de données Python
# str: chaîne de caractères
example_str = "Bonjour, monde!"
print("Type de données str:")
print(type(example_str))  # affiche <class 'str'>

# int: entier
example_int = 42
print("\nType de données int:")
print(type(example_int))  # affiche <class 'int'>

# float: nombre décimal à virgule flottante
example_float = 3.14
print("\nType de données float:")
print(type(example_float))  # affiche <class 'float'>

# complexe: nombre complexe
example_complex = 2 + 3j
print("\nType de données complexe:")
print(type(example_complex))  # affiche <class 'complex'>

# list: liste d'objets
example_list = [1, 2, 3]
print("\nType de données list:")
print(type(example_list))  # affiche <class 'list'>

# tuple: liste immuable d'objets
example_tuple = (1, 2, 3)
print("\nType de données tuple:")
print(type(example_tuple))  # affiche <class 'tuple'>

# range: séquence d'entiers
example_range = range(0, 10)
print("\nType de données range:")
print(type(example_range))  # affiche <class 'range'>

# dict: dictionnaire d'objets
example_dict = {"a": 1, "b": 2, "c": 3}
print("\nType de données dict:")
print(type(example_dict))  # affiche <class 'dict'>

# set/ensemble: ensemble d'objets uniques (non ordonné)
example_set = {1, 2, 3}
print("\nType de données set:")
print(type(example_set))  # affiche <class 'set'>

# frozenset: ensemble d'objets uniques immuable
example_frozenset = frozenset({1, 2, 3})
print("\nType de données frozenset:")
print(type(example_frozenset))  # affiche <class 'frozenset'>

# bool: booléen (vrai ou faux)
example_bool = True
print("\nType de données bool:")
print(type(example_bool))  # affiche <class 'bool'>

# bytes: octets immuables
example_bytes = b"hello"
print("\nType de données bytes:")
print(type(example_bytes))  # affiche <class 'bytes'>

# bytearray: tableau d'octets modifiables
example_bytearray = bytearray(b"hello")
print("\nType de données bytearray:")
print(type(example_bytearray))  # affiche <class 'bytearray'>

# memoryview: vue mémoire pour les objets
example_memoryview = memoryview(b"hello")
print("\nType de données memoryview:")
print(type(example_memoryview))  # affiche <class 'memoryview'>

# NoneType: type de données représentant la valeur None
example_none = None
print("\nType de données NoneType:")
print(type(example_none))  # affiche <class 'NoneType'>
```
